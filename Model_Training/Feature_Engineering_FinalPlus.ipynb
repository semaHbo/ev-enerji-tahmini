```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/ev-enerji-tahmini/train1_scaled_cleaned.csv', index_col='date', parse_dates=True)

print("Satır sayısı:", df.shape[0])
print("Sütun sayısı:", df.shape[1])
df.head()
```

```python
# Gunduz saatlerinde mi
df['GunduzSaatMi'] = df['hour'].apply(lambda x: 1 if 7 <= x <= 19 else 0)

# Sıcaklık farkı İç - Dış
df['Sicaklik_Farki'] = df['Sicaklik_IcOrtalama'] - df['Sicaklik_DisHava']

# Nem farkı İç - Dış
df['Nem_Farki'] = df['Nem_IcOrtalama'] - df['Nem_DisHava']
```

```python
df[['hour', 'GunduzSaatMi', 'Sicaklik_IcOrtalama', 'Sicaklik_DisHava', 'Sicaklik_Farki']].head(10)
```

```python
# Nem farkı İç - Dış
df['Nem_Farki'] = df['Nem_IcOrtalama'] - df['Nem_DisHava']

#  Hafta sonu ve saat etkileşimi
df['IsHaftaSonuVeSaat'] = df['is_weekend'] * df['hour']
```

```python
df[['hour', 'is_weekend', 'IsHaftaSonuVeSaat', 'Nem_IcOrtalama', 'Nem_DisHava', 'Nem_Farki']].head(10)
```

```python

val_df = pd.read_csv('/content/drive/MyDrive/ev-enerji-tahmini/val1_scaled_cleaned.csv', index_col='date', parse_dates=True)

val_df['GunduzSaatMi'] = val_df['hour'].apply(lambda x: 1 if 7 <= x <= 19 else 0)
val_df['Sicaklik_Farki'] = val_df['Sicaklik_IcOrtalama'] - val_df['Sicaklik_DisHava']
val_df['Nem_Farki'] = val_df['Nem_IcOrtalama'] - val_df['Nem_DisHava']
val_df['IsHaftaSonuVeSaat'] = val_df['is_weekend'] * val_df['hour']

test_df = pd.read_csv('/content/drive/MyDrive/ev-enerji-tahmini/test1_scaled_cleaned.csv', index_col='date', parse_dates=True)

test_df['GunduzSaatMi'] = test_df['hour'].apply(lambda x: 1 if 7 <= x <= 19 else 0)
test_df['Sicaklik_Farki'] = test_df['Sicaklik_IcOrtalama'] - test_df['Sicaklik_DisHava']
test_df['Nem_Farki'] = test_df['Nem_IcOrtalama'] - test_df['Nem_DisHava']
test_df['IsHaftaSonuVeSaat'] = test_df['is_weekend'] * test_df['hour']
```

```python
val_df.to_csv('/content/drive/MyDrive/ev-enerji-tahmini/val_plus.csv')
test_df.to_csv('/content/drive/MyDrive/ev-enerji-tahmini/test_plus.csv')
df.to_csv('/content/drive/MyDrive/ev-enerji-tahmini/train_plus.csv')
```
